```python
# For all 328 Pages
counter=1

for i in range(1,328):
    #wait=WebDriverWait(driver,40)
    element=WebDriverWait(driver,40).until(EC.visibility_of_element_located((By.XPATH,'//table[@id="ctl00_ContentPlaceHolder1_GridView1"]')))
 

    container=driver.find_element(By.XPATH,'//table[@id="ctl00_ContentPlaceHolder1_GridView1"]')
    boxes=container.find_elements(By.XPATH,"./tbody/tr")
    
for box in boxes:

    try:
        srno=box.find_element(By.XPATH,"./td[1]/span").text
    except:
        srno=" "


    try:
        proposalNoCheck=box.find_element(By.XPATH,"./td[2]/table/tbody/tr[1]/td[1]/b").text
        #print(proposalNoCheck)

        if proposalNoCheck == "Proposal No":
            proposalNumber = box.find_element(By.XPATH,'./td[2]/table/tbody/tr[1]/td[3]/span').text
            #print(proposalNumber)

    except:
        proposalNumber=" "
        continue

    try:
        fileNoCheck=box.find_element(By.XPATH,'./td[2]/table/tbody/tr[2]/td[1]/b').text
        #print(fileNoCheck)

        if fileNoCheck== "File No":
            fileNo= box.find_element(By.XPATH,'./td[2]/table/tbody/tr[2]/td[3]/span').text
            #print(fileNo)
    except:
        fileNo=" "


        
    try:
        proposalNameCheck=box.find_element(By.XPATH,'./td[2]/table/tbody/tr[3]/td[1]/b').text
        #print(proposalNameCheck)

        if proposalNameCheck== "Proposal Name":
            proposalName= box.find_element(By.XPATH,'./td[2]/table/tbody/tr[3]/td[3]/span').text
            #print(proposalName)

    except:
        proposalName=" "

 

        
    try:
        CategoryCheck=box.find_element(By.XPATH,'./td[2]/table/tbody/tr[4]/td[1]/b').text
        #print(CategoryCheck)

        if CategoryCheck== "Category":
            Category= box.find_element(By.XPATH,'./td[2]/table/tbody/tr[4]/td[3]/span').text
            #print(Category)

    except:
        Category=" "
 


    try:
        StateCheck=box.find_element(By.XPATH,'./td[3]/table/tbody/tr[1]/td[1]/b').text
        #print(StateCheck)

        if StateCheck== "State":
            State= box.find_element(By.XPATH,'./td[3]/table/tbody/tr[1]/td[3]/span[2]').text
            #print(State)
    except:
        State=" "

    try:
        DistrictCheck=box.find_element(By.XPATH,'./td[3]/table/tbody/tr[2]/td[1]/b').text
        #print(DistrictCheck)

        if DistrictCheck== "District":
            District= box.find_element(By.XPATH,'./td[3]/table/tbody/tr[2]/td[3]/span[2]').text
            #print(District)

    except:
        District=" "
        

    try:
        TehsilCheck=box.find_element(By.XPATH,'./td[3]/table/tbody/tr[3]/td[1]/b').text
        #print(TehsilCheck)

        if TehsilCheck== "Tehsil":
            Tehsil= box.find_element(By.XPATH,'./td[3]/table/tbody/tr[3]/td[3]/span[2]').text
            #print(Tehsil)

    except:
        Tehsil=" "

        
          
        
    try:
        DateOfApplicationCheck=box.find_element(By.XPATH,"./td[4]/span").text
    except:
        DateOfApplicationCheck=" "
        
        
        

    try:
        CompanyName=box.find_element(By.XPATH,'./td[5]/span').text
        #print(CompanyName)

    except:
        CompanyName=" "   


    try:
        projectType=box.find_element(By.XPATH,'./td[6]/span').text
        #print(projectType)

    except:
        projectType=" "   

 
    try:
        DateDetails=box.find_element(By.XPATH,'./td[11]/a').get_attribute('href')
        #print(DateDetails)

    except:
        DateDetails=" "   
       
    
    

    dfmain=pd.DataFrame({'Proposal_No':[proposalNumber],'File_No':[fileNo],'Proposal_name':[proposalName],'Category':[Category],'State':[State],'District':[District],'Tehsil':[Tehsil],
                  'Project_Type':[projectType],'Date_Of_Application':[DateOfApplicationCheck],'Company_Name':[CompanyName],'project_Type':[projectType],'Date_Details':[DateDetails], 'srno':[srno]})

    
    if dfcounter==0:
        dfmain.to_csv(r"D:\OneDrive - Indian School of Business\Aditya_ShekharSir_Env_Clearance_Project\Env_Clearance_Project\Data\Raw Data\_2.csv", index=False)
        dfcounter=1
    else:
        dfmain.to_csv(r"D:\OneDrive - Indian School of Business\Aditya_ShekharSir_Env_Clearance_Project\Env_Clearance_Project\Data\Raw Data\_2.csv", index=False, mode='a', header=False)

```
